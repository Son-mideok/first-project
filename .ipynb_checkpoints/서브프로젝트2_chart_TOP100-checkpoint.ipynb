```python
# 비동기,제이슨

import requests
from bs4 import BeautifulSoup

BASE_URL = 'https://api.card-gorilla.com:8080/v1/charts/ranking?'
parameters = 'date=2022-09-07&term=weekly&card_gb=CRD&limit=100&chart=top100&idx=&idx2='


URL = BASE_URL + parameters
```

```python
res = requests.get(URL)
res.status_code
```

```python
import json
json_data = json.loads(res.text)
```
